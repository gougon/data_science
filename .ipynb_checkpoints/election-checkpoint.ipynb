```python
import pyecharts
import pandas as pd
import pyecharts.options as opts
import re
from pyecharts.charts import Line
from pyecharts.globals import ThemeType
from pandas import DataFrame
from sklearn import preprocessing
```

```python
electionDF = pd.read_csv('election.csv',encoding = 'ANSI')

date = list(electionDF.loc[:, 'Date'])

## 使用標準化Z-Score套件
z_score_scaler = preprocessing.MinMaxScaler(feature_range = (0,1))
## 對數據進行標準化
election_z_score =  pd.DataFrame(z_score_scaler.fit_transform(electionDF.drop(['Date'], axis = 1)),columns=['Trump','Biden'])

```

```python

line = Line(init_opts=opts.InitOpts(theme=ThemeType.CHALK))
  
line.add_xaxis(xaxis_data=date).add_yaxis(
        series_name="Trump",
        y_axis=election_z_score['Trump'],
        label_opts=opts.LabelOpts(is_show=False),
    ).add_yaxis(
        series_name="Biden",
        y_axis=election_z_score['Biden'],
        label_opts=opts.LabelOpts(is_show=False),
    ).set_global_opts(
        title_opts=opts.TitleOpts(title="總統候選人網路聲量"),
        tooltip_opts=opts.TooltipOpts(trigger="axis"),
        yaxis_opts=opts.AxisOpts(
            type_="value",
            axistick_opts=opts.AxisTickOpts(is_show=True),
            splitline_opts=opts.SplitLineOpts(is_show=True),
        ),
        xaxis_opts=opts.AxisOpts(type_="category", boundary_gap=False),
    ).render_notebook()
```

```python
youtubeDF = pd.read_csv('US_youtube_trending_data.csv')  
keys = list(youtubeDF.keys())
keys.remove('trending_date')

trendingDate=youtubeDF.drop_duplicates(subset='trending_date', keep='first', inplace=False).drop(keys,axis=1)
keys.remove('view_count')
TrumpViewCount=trendingDate.merge(youtubeDF[youtubeDF['title'].str.contains("Trump", case=False,regex=False)].drop(keys,axis=1).groupby(youtubeDF['trending_date']).sum().reset_index(),on='trending_date', how='left').fillna(0)
BidenViewCount=trendingDate.merge(youtubeDF[youtubeDF['title'].str.contains("Biden", case=False,regex=False)].drop(keys,axis=1).groupby(youtubeDF['trending_date']).sum().reset_index(),on='trending_date', how='left').fillna(0)

totalViewCount = BidenViewCount.merge(TrumpViewCount, how='inner', on=['trending_date'],suffixes=('_Biden','_Trump'))
## 對數據進行標準化
ViewCount_z_score =  pd.DataFrame(z_score_scaler.fit_transform(totalViewCount.drop(['trending_date'], axis = 1)),columns=['view_count_Biden','view_count_Trump'])
print(ViewCount_z_score)
```

```python
youtubeLine = Line(init_opts=opts.InitOpts(theme=ThemeType.CHALK))
  
youtubeLine.add_xaxis(xaxis_data=date).add_yaxis(
        series_name="Trump View Count",
        y_axis=ViewCount_z_score['view_count_Trump'],
        label_opts=opts.LabelOpts(is_show=False),
        is_smooth=True,
    ).add_yaxis(
        series_name="Biden View Count",
        y_axis=ViewCount_z_score['view_count_Biden'],
        label_opts=opts.LabelOpts(is_show=False),
         is_smooth=True
    
    ).add_yaxis(
        series_name="Trump",
        y_axis=election_z_score['Trump'],
        label_opts=opts.LabelOpts(is_show=False),
    ).add_yaxis(
        series_name="Biden",
        y_axis=election_z_score['Biden'],
        label_opts=opts.LabelOpts(is_show=False),
    ).set_global_opts(
        title_opts=opts.TitleOpts(title="Youtube Trends View Count"),
        tooltip_opts=opts.TooltipOpts(trigger="axis"),
        yaxis_opts=opts.AxisOpts(
            type_="value",
            axistick_opts=opts.AxisTickOpts(is_show=True),
            splitline_opts=opts.SplitLineOpts(is_show=True),
        ),
        xaxis_opts=opts.AxisOpts(type_="category", boundary_gap=False),
    ).render_notebook()
```
